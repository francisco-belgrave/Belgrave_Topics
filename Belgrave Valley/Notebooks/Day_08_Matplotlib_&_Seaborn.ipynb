### Line Charts

We will use this [csv](Resources/unrate.csv)

```python
unrate = pd.read_csv('Resources/unrate.csv')
unrate.DATE = pd.to_datetime(unrate.DATE)
unrate.head(12)
```

You can do `shift` + `tab` inside the brackets of a function to access more information about it.

```python
plt.plot(unrate.DATE.head(12), unrate.VALUE.head(12), 'r*-')
plt.xticks(rotation = 90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```

```python
fig = plt.figure(figsize = (14, 20))
plt.title('Monthly Unemployment Trends, 1948-1952')

ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)

ax1.plot(unrate.DATE.head(12), unrate.VALUE.head(12),'*-')
# ax1.set_title('Monthly Unemployment Rate, 1948')

ax2.plot(unrate.DATE[12:24],unrate.VALUE[12:24],'r*-')
# ax2.set_title('Monthly Unemployment Rate, 1949')

ax3.plot(unrate.DATE[24:36],unrate.VALUE[24:36],'y*-')
# ax3.set_title('Monthly Unemployment Rate, 1950')

ax4.plot(unrate.DATE[36:48],unrate.VALUE[36:48],'g*-')
# ax4.set_title('Monthly Unemployment Rate, 1951')

ax5.plot(unrate.DATE[48:60],unrate.VALUE[48:60],'m*-')
# ax5.set_title('Monthly Unemployment Rate, 1952')

plt.show()
```

Using functions rather than methods of axis and figures

```python
plt.figure(figsize = (14, 20))

plt.subplot(5, 1, 1)
plt.plot(unrate.DATE.head(12), unrate.VALUE.head(12),'*-')
plt.title('Monthly Unemployment Rate, 1948')

plt.subplot(5, 1, 2)
plt.plot(unrate.DATE[12:24], unrate.VALUE[12:24],'r*-')
plt.title('Monthly Unemployment Rate, 1949')

plt.subplot(5, 1, 3)
plt.plot(unrate.DATE[24:36], unrate.VALUE[24:36],'y*-')
plt.title('Monthly Unemployment Rate, 1950')

plt.subplot(5, 1, 4)
plt.plot(unrate.DATE[36:48], unrate.VALUE[36:48],'g*-')
plt.title('Monthly Unemployment Rate, 1951')

plt.subplot(5, 1, 5)
plt.plot(unrate.DATE[48:60], unrate.VALUE[48:60],'m*-')
plt.title('Monthly Unemployment Rate, 1952')

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

plt.figure(figsize = (6, 3))

plt.title('Monthly Unemployment Rate, 1948-1949')

plt.plot(unrate.MONTH.head(12), unrate.VALUE.head(12),'r')
plt.plot(unrate.MONTH[12:24], unrate.VALUE[12:24],'b')

plt.xlabel('Month')
plt.ylabel('Unemployment Rate')

plt.show()
```

```python
plt.figure(figsize = (10, 6))

plt.title('Monthly Unemployment Trends, 1948-1952')

plt.plot(unrate.MONTH.head(12),unrate.VALUE.head(12),'r')
plt.plot(unrate.MONTH[12:24],unrate.VALUE[12:24],'b')
plt.plot(unrate.MONTH[24:36],unrate.VALUE[24:36],'g')
plt.plot(unrate.MONTH[36:48],unrate.VALUE[36:48],'orange')
plt.plot(unrate.MONTH[48:60],unrate.VALUE[48:60],'k')

plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')

plt.legend(('1948','1949','1950','1951','1952'), loc ='upper left')

plt.show()
```

### Scatter Plots

For this section we will use a movies rating [csv](Resources/fandango_scores.csv)

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

fig3 = plt.figure(figsize=(20, 12))

for sp in range(0, 4):
    
    ax = fig3.add_subplot(2,2, sp+1)
    
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[major_cats[sp]], c='green', label='Men')
    
    plt.xlim(1968, 2011)
    plt.ylim(0, 100)
    
    plt.xlabel('Year')
    plt.ylabel('%')
    plt.title(f'{major_cats[sp]}') 
    plt.legend(loc='upper center')
    
    ax.axhline(50, linestyle='--', color='k', alpha=0.15) # You can also use plt.hlines()
    
    ax.tick_params(bottom =False, top =False,
                 left =False, right =False)

    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

plt.show()
```

```python
stem_cats = ['Engineering', 'Computer Science',
             'Psychology', 'Biology',
             'Physical Sciences', 'Math and Statistics']

cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

fig3 = plt.figure(figsize = (20, 4))
end = 6

for sp in range(0, end):
    ax = fig3.add_subplot(1,6,sp+1)
    
    if sp == 0:
        ax.text(2005, 87,"Men")
        ax.text(2002, 8,"Women")
    
    if sp == end - 1:
        ax.text(2005, 62,"Men")
        ax.text(2001, 35,"Women")
        
    ax.plot(women_degrees['Year'],
            women_degrees[stem_cats[sp]],
            c = cb_dark_blue, 
            label='Women',
            linewidth=4)
    
    ax.plot(women_degrees['Year'],
            100-women_degrees[stem_cats[sp]],
            c=cb_orange, 
            label='Men', 
            linewidth=4)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    if sp > 0:
        ax.get_yaxis().set_visible(False)
        
    ax.axhline(50, linestyle='--', color='k', alpha=0.12)        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.show()
```

```python
lib_arts_cats = ['Foreign Languages', 'English',
                 'Communications and Journalism',
                 'Art and Performance',
                 'Social Sciences and History']

other_cats = ['Health Professions',
              'Public Administration',
              'Education','Agriculture','Business',
              'Architecture']

fig2 = plt.figure(figsize = (16, 20))

# First column: STEM degrees.

for sp in range(0,18,3):
    cat_index = int(sp/3)
    
    ax = fig2.add_subplot(6, 3, sp+1)
    
    ax.plot(women_degrees['Year'],
            women_degrees[stem_cats[cat_index]],
            c = cb_dark_blue, label ='Women',
            linewidth = 3)
    ax.plot(women_degrees['Year'],
            100-women_degrees[stem_cats[cat_index]],
            c = cb_orange, label ='Men', linewidth = 3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), 
               alpha = 0.5)  
    ax.tick_params(bottom = False, top = False ,
                   left = False, right = False,
                   labelbottom = False)
    
    if cat_index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom = True)
       
# Second column: Liberal Arts degrees.

for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    
    ax = fig2.add_subplot(6,3,sp+1)
    
    ax.plot(women_degrees['Year'],
            women_degrees[lib_arts_cats[cat_index]],
            c = cb_dark_blue, label ='Women',
            linewidth=3)
    ax.plot(women_degrees['Year'],
            100-women_degrees[lib_arts_cats[cat_index]],
            c = cb_orange, label ='Men', linewidth = 3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255),
               alpha = 0.5)
    ax.tick_params(bottom = False, top = False ,
                   left = False, right = False,
                   labelbottom = False)
    
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif cat_index == 4:
        ax.tick_params(labelbottom = True)

# Third column: Other degrees.

for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    
    ax = fig2.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'],
            women_degrees[other_cats[cat_index]],
            c = cb_dark_blue, label ='Women', 
            linewidth = 3)
    ax.plot(women_degrees['Year'], 
            100-women_degrees[other_cats[cat_index]],
            c = cb_orange, label ='Men', linewidth = 3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255),
               alpha = 0.5)
    ax.tick_params(bottom = False, top = False ,
                   left = False, right = False,
                   labelbottom = False)
    
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom = True)

plt.savefig("gender_degrees.png")
plt.show()
```

###### Processing

```python
def wage_split(x):
    try:
        return int(x.split("K")[0][1:])
    except:
        return 0
player_df['Wage'] = player_df['Wage'].apply(lambda x : wage_split(x))

def value_split(x):
    try:
        if 'M' in x:
            return float(x.split("M")[0][1:])
        elif 'K' in x:
            return float(x.split("K")[0][1:])/1000
    except:
        return 0
    
player_df['Value'] = player_df['Value'].apply(lambda x : value_split(x))
```

```python
numcols = [
 'Overall',
 'Potential',
'Crossing','Finishing',  'ShortPassing',  'Dribbling','LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility',  'Stamina',
 'Value','Wage']
catcols = ['Name','Club','Nationality','Preferred Foot','Position','Body Type']
```

```python
player_df = player_df[numcols+catcols]
```

```python
filtered_player_df = player_df[(player_df['Club'].isin(['FC Barcelona', 'Paris Saint-Germain',
       'Manchester United', 'Manchester City', 'Chelsea', 'Real Madrid','FC Porto','FC Bayern München'])) & 
                      (player_df['Nationality'].isin(['England', 'Brazil', 'Argentina',
       'Brazil', 'Italy','Spain','Germany'])) 
                     ]

df_plot = filtered_player_df.groupby(['Club', 'Nationality']).size().reset_index().pivot(columns='Club', index='Nationality', values=0).reset_index()
```

```python
max_wage = filtered_player_df.Wage.max()
max_wage_player = filtered_player_df[(player_df['Wage'] == max_wage)]['Name'].values[0]
```
