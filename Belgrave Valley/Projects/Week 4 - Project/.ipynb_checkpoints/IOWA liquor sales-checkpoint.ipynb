# Getting started

Once you've chosen your scenario, download the data from [the Iowa website](https://data.iowa.gov/Economy/Iowa-Liquor-Sales/m3tr-qhgy) in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

```python
from __future__ import division
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
## Load the data into a DataFrame
stores_df = pd.read_csv('../Week 4 - Project/Iowa_Liquor_Sales.csv')
stores_df.head()
```

```python
#Drop NaN values
stores_clean = stores_df.dropna(axis=0)

# Transform the dates to datetime format
stores_clean.loc[:, "Date"] = pd.to_datetime(stores_clean.loc[:, 'Date'], format='%m/%d/%Y')

# Rename columns to remove whitespaces
stores_clean = stores_clean.rename(columns = {i: i.replace (" ", "_") for i in stores_clean.columns})

# Set category columns to category types
stores_clean.loc[:, "Store_Number"] = stores_clean.loc[:, "Store_Number"].astype('category')
stores_clean.loc[:, "County_Number"] = stores_clean.loc[:, "County_Number"].astype('category')
stores_clean.loc[:, "Category"] = stores_clean.loc[:, "Category"].astype('category')
stores_clean.loc[:, "Category_Name"] = stores_clean.loc[:, "Category_Name"].astype('category')
stores_clean.loc[:, "Vendor_Number"] = stores_clean.loc[:, "Vendor_Number"].astype('category')
stores_clean.loc[:, "Item_Number"] = stores_clean.loc[:, "Item_Number"].astype('category')

#Set prices columns to floats
for col in ["State_Bottle_Cost", "State_Bottle_Retail", "Sale_(Dollars)"]:
    stores_clean.loc[:, col] = stores_clean.loc[:, col].apply(lambda x: x.replace("$", ""))
    stores_clean.loc[:, col] = stores_clean.loc[:, col].astype('float')

stores_clean.info()
```

# Mine the data
Now you are ready to compute the variables you will use for your regression from the data. For example, you may want to

```python
#Get year and month of each sale in a separate column
stores_clean['year'] = stores_clean['Date'].apply(lambda x: x.year)
stores_clean['month'] = stores_clean['Date'].apply(lambda x: x.month)
stores_clean.head()
```

```python
#Orders in 2015
stores_2015 = stores_clean.loc[stores_clean['year']==2015, ['Store_Number', 'Zip_Code', 'County', 'City', 'Vendor_Number','Sale_(Dollars)', 'Volume_Sold_(Liters)', 'Bottles_Sold', 'State_Bottle_Retail', 'month']]
stores_2015.head()
```

```python
#Total sales per store per month in 2015
stores_stats_2015 = pd.pivot_table(stores_2015, index=['Store_Number'], columns=['month'], values=['Sale_(Dollars)'], aggfunc=sum, margins=True, margins_name='Yearly_Sales')

stores_stats_2015.columns = stores_stats_2015.columns.droplevel()

stores_stats_2015 = stores_stats_2015.drop(range(4,13), axis=1)

stores_stats_2015 = stores_stats_2015.rename(columns={1: 'Month_1_Sales', 2: 'Month_2_Sales', 3: 'Month_3_Sales'})

stores_stats_2015.head()
```

```python
#Add volumes solds per store in 2015
vols_sold_2015 = pd.pivot_table(stores_2015, index=['Store_Number'], values=['Volume_Sold_(Liters)'], columns=['month'], aggfunc=np.sum, margins=True, margins_name='Yearly_Volumes')

vols_sold_2015.columns = vols_sold_2015.columns.droplevel()

vols_sold_2015 = vols_sold_2015.drop(range(4,13), axis=1)

vols_sold_2015 = vols_sold_2015.rename(columns={1: 'Month_1_Volumes', 2: 'Month_2_Volumes', 3: 'Month_3_Volumes'})

stores_stats_2015 = stores_stats_2015.join(vols_sold_2015, how='left')

stores_stats_2015.head()
```

```python
#Add bottles sold per store in 2015
bottles_sold_2015 = pd.pivot_table(stores_2015, index=['Store_Number'], values=['Bottles_Sold'], columns=['month'], aggfunc=np.sum, margins=True, margins_name='Yearly_Bottles')

bottles_sold_2015.columns = bottles_sold_2015.columns.droplevel()

bottles_sold_2015 = bottles_sold_2015.drop(range(4,13), axis=1)

bottles_sold_2015 = bottles_sold_2015.rename(columns={1: 'Month_1_Bottles', 2: 'Month_2_Bottles', 3: 'Month_3_Bottles'})

stores_stats_2015 = stores_stats_2015.join(bottles_sold_2015, how='left')

stores_stats_2015.head()
```

```python
#Add average price per bottle
price_per_bottle_2015 = pd.pivot_table(stores_2015, index=['Store_Number'], values=['State_Bottle_Retail'], aggfunc=np.mean)
price_per_bottle_2015 = price_per_bottle_2015.rename(columns={'State_Bottle_Retail': 'Average_Bottle_Price'})

stores_stats_2015 = stores_stats_2015.join(price_per_bottle_2015, how='left')

stores_stats_2015 = stores_stats_2015.dropna()
stores_stats_2015.head()
```

```python
#Orders in 2016
stores_2016 = stores_clean.loc[stores_clean['year']==2016, ['Store_Number', 'Zip_Code', 'County', 'City', 'Vendor_Number','Sale_(Dollars)', 'Volume_Sold_(Liters)', 'Bottles_Sold', 'State_Bottle_Retail', 'month']]
stores_2016.head()
```

```python
#Total sales per store per month in 2016
stores_stats_2016 = pd.pivot_table(stores_2016, index=['Store_Number'], columns=['month'], values=['Sale_(Dollars)'], aggfunc=sum, margins=True, margins_name='Yearly_Sales')

stores_stats_2016.columns = stores_stats_2016.columns.droplevel()

stores_stats_2016 = stores_stats_2016.rename(columns={1: 'Month_1_Sales', 2: 'Month_2_Sales', 3: 'Month_3_Sales'})

stores_stats_2016.head()
```

```python
#Add volumes solds per store in 2015
vols_sold_2016 = pd.pivot_table(stores_2016, index=['Store_Number'], values=['Volume_Sold_(Liters)'], columns=['month'], aggfunc=np.sum, margins=True, margins_name='Yearly_Volumes')

vols_sold_2016.columns = vols_sold_2016.columns.droplevel()

vols_sold_2016 = vols_sold_2016.rename(columns={1: 'Month_1_Volumes', 2: 'Month_2_Volumes', 3: 'Month_3_Volumes'})

stores_stats_2016 = stores_stats_2016.join(vols_sold_2016, how='left')

stores_stats_2016.head()
```

```python
#Add bottles sold per store in 2016
bottles_sold_2016 = pd.pivot_table(stores_2016, index=['Store_Number'], values=['Bottles_Sold'], columns=['month'], aggfunc=np.sum, margins=True, margins_name='Yearly_Bottles')

bottles_sold_2016.columns = bottles_sold_2016.columns.droplevel()

bottles_sold_2016 = bottles_sold_2016.rename(columns={1: 'Month_1_Bottles', 2: 'Month_2_Bottles', 3: 'Month_3_Bottles'})

stores_stats_2016 = stores_stats_2016.join(bottles_sold_2015, how='left')

stores_stats_2016.head()
```

```python
#Add average price per bottle
price_per_bottle_2016 = pd.pivot_table(stores_2016, index=['Store_Number'], values=['State_Bottle_Retail'], aggfunc=np.mean)
price_per_bottle_2016 = price_per_bottle_2015.rename(columns={'State_Bottle_Retail': 'Average_Bottle_Price'})

stores_stats_2016 = stores_stats_2016.join(price_per_bottle_2016, how='left')

stores_stats_2016 = stores_stats_2016.dropna()
stores_stats_2016.head()
```

# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

### Model 1: Linear regression with sales, Bottles Sold and Volume Sold  from January to March

```python
from sklearn import linear_model
from sklearn.cross_validation import train_test_split

#Build linear model to predict sales based on data from 2015
X_2015 = stores_stats_2015[['Month_1_Sales', 'Month_2_Sales', 'Month_3_Sales', 'Month_1_Bottles', 'Month_2_Bottles', 'Month_3_Bottles', 'Month_1_Volumes', 'Month_2_Volumes', 'Month_3_Volumes']]
y_2015 = stores_stats_2015['Yearly_Sales']

X_train, X_test, y_train, y_test = train_test_split(X_2015, y_2015, test_size = 0.2)

lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)

print 'Model coefficients:', lm.coef_
```

```python
#Test model on test data
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

predictions_test = lm.predict(X_test)

print 'Mean squarred error:', mean_squared_error(y_test, predictions_test)
print 'Root mean squarred error:', np.sqrt(mean_squared_error(y_test, predictions_test))
print 'Mean absolute error:', mean_absolute_error(y_test, predictions_test)
print 'r^2 score:', r2_score(y_test, predictions_test)
```

r^2 score is high which suggests that the model is a good fit. However, the mean squarred error is also very high so we will try fitting a Lasso regression to the data instead to see if some parameters could be ignored in the model.

### Model 2: Linear model with Lasso regression 

```python
#Linear model with Lasso regression
llm = linear_model.Lasso(alpha=5)
llm.fit(X_train, y_train)

print 'Model with Lasso coefficients:', llm.coef_
print '-----------'

predictions_lasso = llm.predict(X_test)

print 'Mean squarred error:', mean_squared_error(y_test, predictions_lasso)
print 'Root mean squarred error:', np.sqrt(mean_squared_error(y_test, predictions_lasso))
print 'Mean absolute error:', mean_absolute_error(y_test, predictions_lasso)
print 'r^2 score:', r2_score(y_test, predictions_lasso)
```

### Cross Validation of model 1

```python
#Evaluate model fit with 5-fold cross-validation
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(lm, X_2015, y_2015, cv=5)

print 'Cross-validation scores:', scores
print 'Mean cross-validation score:', scores.mean()
```

### Cross Validation of model 2

```python
#Evaluate model fit with 5-fold cross-validation
from sklearn.cross_validation import cross_val_score

scores_lasso = cross_val_score(llm, X_2015, y_2015, cv=5)

print 'Cross-validation scores:', scores_lasso
print 'Mean cross-validation score:', scores_lasso.mean()
```

Cross-validation r^2 scores are high so our model seems to be a good fit.  
The Lasso regression did not improve the errors or r^2 scores so we will our initial linear model to predict sales values in 2016.
