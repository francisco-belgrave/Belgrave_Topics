## Regular Expressions

#### 1. Write a function to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

````python
import re

def is_allowed_specific_char(string):
    charRe = re.compile(r'[^a-zA-Z0-9.]')
    string = charRe.search(string)
    return not bool(string)

print(is_allowed_specific_char("ABCDEFabcdef123450")) 
print(is_allowed_specific_char("*&%@#!}{"))
````

#### 2. Write a function that matches a string that has an `a` followed by zero or more b's.

````python
import re

def text_match(text):
        patterns = 'ab*?'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("zab"))
print(text_match("abc"))
print(text_match("abbc"))
````

#### 3. Write a Python program to find sequences of lowercase letters joined with a underscore.

````python
import re

def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc"))
print(text_match("aab_Abbbc"))
print(text_match("Aaab_abbbc"))
````

#### 8. convert to 'v' if a character is a vowel and 'c' if a consonant, otherwise convert to '?'

I've conveniently defined the vowels and lowercase alphabet below. The characters to convert are in the `characters` list.

````python
alphabet = list(string.ascii_lowercase)
vowels = list('aeiou')
print (alphabet)
print (vowels)

characters = ['a','f',None,'k','l','1',12,'e','e',-1,'i','b','p']
````

#### 8.1 With a for-loop

````python
vc =[]
for c in characters:
    
    if c in vowels:
        vc.append('v')
    elif c in alphabet:
        vc.append('c')
    else:
        vc.append('?')
        
vc
````

#### 8.2 With a list comprehension

````python
vc = ['c' if car in alphabet else 'v' if car in vowels else '?' for car in characters]
vc
````

````python
print (['v' if ch in vowels else 'c' if ch in alphabet else '?' for ch in characters])
````
